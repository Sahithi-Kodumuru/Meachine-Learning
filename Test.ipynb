```python
#task-2
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'OrderDate': pd.to_datetime(['2025-03-01', '2025-03-10', '2025-03-20']),
    'Amount': [100, 200, 300]
})
df
```

```python
#extract the month and year from the order date
df['Month'] = df['OrderDate'].dt.month
df['Year'] = df['OrderDate'].dt.year
print(df)
```

```python
#calculate the number of days between today and each order date
from datetime import datetime
today = datetime.now()
df['DaysBetween'] = (today - df['OrderDate']).dt.days
df
```

```python
import pandas as pd                  # 28-03-2025
import numpy as np
alcohol = pd.read_csv("C:\\Users\\kodum\\Downloads\\data.txt")
alcohol
```

```python
#1 which continent drinks more beer on averge
avg_cons = alcohol.groupby('continent')['beer_servings'].mean()
avg_cons
```

```python
#2 For each continent determine the statistics for wine consumption
statistics = alcohol.groupby('continent')['wine_servings'].agg(
    mean = np.mean,
    median = np.median,
    std = np.std,
    min = np.min,
    max = np.max
)
statistics    
```

```python
#3 Calculate the avg alcohol consumption for each continent based on the available data
avg_alc_cons = alcohol.groupby('continent')['total_litres_of_pure_alcohol'].mean()
avg_alc_cons
```

```python
#4 Calculate the median of alcohol consumption for each continent across all columns of the dataset
median_stats = alcohol.groupby('continent')[['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol']].median()
median_stats
```

```python
#5 Display the mean or average minimum and maximum values for the sprit consumption(lowest and the highest values)
# min values
min_values = alcohol.groupby('continent').spirit_servings.idxmin()  # alcohol.groupby('continent')['spirit_servings'].max()
min_values
```

```python
# max values
max_values = alcohol.groupby('continent')['spirit_servings'].idxmax()
max_values
```

```python
#3 Calculate the avg alcohol consumption for each continent based on the available data
avg = alcohol.groupby('continent').mean(numeric_only=True)
avg
```
